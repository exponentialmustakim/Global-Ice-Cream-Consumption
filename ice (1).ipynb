```python
#importing the relevant libraries 
import pandas as pd
import numpy as np
```

```python
#loading the dataset into dataframe
ic = pd.read_csv(r'R skills/R skills/Input/ice.cream.data.csv')
ic
```

```python
# Removal of unnecessary columns
ic=ic[['Region','Country', 'Data Type','Unit', 'Unit Multiplier','2000', '2001', '2002',
       '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011',
       '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']]
ic.head(5)
```

```python
#converting the data into relevant units by multipling the relevant data columns with it's respective  Unit Multiplier
ic.loc[:,'2000':'2019'] =ic.loc[:,'2000':'2019'].multiply(ic['Unit Multiplier'], axis=0)
ic.head(5)
```

```python
# Removal of 'Unit' and 'Unit Multiplier' columns as they are not necessary for further analysis
ic.drop(['Unit','Unit Multiplier'],axis=1,inplace=True)
ic.head(5)
```

```python
#converting the year columns into a single row and putting their value in reading column 
icm = pd.melt(ic, id_vars=['Country', 'Region','Data Type'],
      var_name="Year",
     value_name="reading")
icm
```

```python
#pivoting the data 
icm=icm.pivot(columns="Data Type", values="reading", index=['Region','Country','Year']).reset_index()
```

```python
icm
```

```python
# removing comumn names
icm.columns.name = None
icm
```

```python
#Removing unnecessary columns
icm.drop(['Retail Volume (Tonnes)','Total Volume','Total Volume (Tonnes)'],axis=1,inplace=True)
icm.head(30)
```

```python
#Renaming columns 'Retail Value RSP' to 'Ice Cream Value', 'Retail Volume' to 'Ice Cream Volume'
icm.rename(columns={'Retail Value RSP': 'Ice Cream Value', 'Retail Volume': 'Ice Cream Volume'}, inplace=True)
icm.head(5)
```

```python
#creating 'Ice Cream Price' column 
icm['Ice Cream Price'] = icm['Ice Cream Value']/icm['Ice Cream Volume']
icm
```

```python
#For each year, total volume by Region
icm.groupby(['Region','Year'])['Ice Cream Volume'].sum().reset_index()
```

```python
md = pd.read_csv(r'R skills/R skills/Input/macro.data.csv')
md
```

```python
md.columns
```

```python
md=md[['Region','Country', 'Subcategory','Unit', 'Unit Multiplier','2000', '2001', '2002',
       '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011',
       '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']]
md.head(5)
```

```python
md.loc[:,'2000':'2019'] =md.loc[:,'2000':'2019'].multiply(md['Unit Multiplier'], axis=0)
md.head(5)
```

```python
md.drop(['Unit','Unit Multiplier'],axis=1,inplace=True)
md.head(5)
```

```python
mdd = pd.melt(md, id_vars=['Country', 'Region','Subcategory'],
      var_name="Year",
     value_name="reading")
mdd
```

```python
mdd=mdd.pivot(columns="Subcategory", values="reading", index=['Region','Country','Year']).reset_index()
```

```python
mdd.columns.name = None
mdd
```

```python
mdd.columns
```

```python

mdd.rename(columns={'  Total GDP': 'GDP', 'Population: National Estimates at January 1st': 'Population'}, inplace=True)
mdd.head(5)
```

```python

df = pd.merge(icm, mdd, left_on=['Country', 'Year'], right_on=['Country', 'Year'], how='left')
```

```python
df
```

```python
# prompt: 3.	Calculate per capita variables where appropriate.

df['Ice Cream Per Capita'] = df['Ice Cream Volume'] / df['Population']
df['GDP Per Capita'] = df['GDP'] / df['Population']

```

```python
df.sample(5)
```

```python
#check if all the values of 	Region_x is equal to 	Region_y

(df['Region_x'] == df['Region_y']).all()
```

```python
df.drop('Region_y',axis=1,inplace=True)
```

```python
df.rename(columns={'Region_x':'Region'},inplace=True)
```

```python
df
```

```python
# 	Visualize the relationship between Ice Cream Volume per Capita and GDP per Capita

import matplotlib.pyplot as plt
import seaborn as sns
sns.lmplot(data=df,x='Ice Cream Per Capita', y='GDP Per Capita')
plt.xlabel('Ice Cream Per Capita')
plt.ylabel('GDP Per Capita')

plt.savefig('R skills/R skills/Output/icecream_gdp_scatter.png',dpi=300)
plt.show()
```

```python


def percent_missing(df):
  missing_values = df.isnull().sum()
  total_values = df.shape[0] * df.shape[1]
  missing_percent = missing_values / total_values * 100
  return missing_percent

missing_values_df = percent_missing(df)
print(missing_values_df)

```
